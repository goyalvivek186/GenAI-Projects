```python
from openai import OpenAI
import os
```

```python
groq = OpenAI(api_key=os.getenv('GROQ_API_KEY'), base_url=os.getenv('base_url'))
```

```python
common_prompt = 'You are one of 3 friends having a conversation. '
```

```python
system_prompts ={
    'Aron' : 'You are Aron friendly and incouraging in nature. You try to solve any situation without escalating. reply as yourself in 2 sentences max only',
    'Blake' : 'You are Blake arrogant and sarcastic friend who is know it all and proud. reply as yourself in 2 sentences max only',
    'Charly' : 'You are Charly cunning friend who likes to take advantage of a situation by esclating it and making a mess not on your name but others. reply as yourself in 2 sentences max only'
}
```

```python
for friend,prompt in system_prompts.items():
    print(friend + ' ' + prompt)
```

```python
user_prompts = '''conversation history - 
Charley- Hi
'''
```

```python
friends = ['Aron', 'Blake', 'Charly']
```

```python
def append_messages(friend, msg):
    global user_prompts
    user_prompts += f'{friend}- {msg}\n'
```

```python
def create_groq_message(system_prompt, user_prompt):
    return[{
        'role' : 'system',
        'content' : system_prompt
    },
    {
        'role' : 'user',
        'content' : user_prompt
    }
    ]
```

```python
def chat():
    for i in range(1):
        for friend,system_prompt in system_prompts.items():
            response = groq.chat.completions.create(messages= create_groq_message(system_prompt, user_prompts), model='llama-3.1-8b-instant')
            append_messages(friend, response.choices[0].message.content)
            # return friend + '-' + response.choices[0].message.content
    
    return user_prompts
    
```

```python
print(chat())
```
