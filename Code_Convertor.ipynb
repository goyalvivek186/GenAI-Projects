```python
from openai import OpenAI
from google.colab import userdata
import gradio as gr
```

```python
groq = OpenAI(api_key=userdata.get('GROQ_API_KEY'), base_url=userdata.get('GROQ_BASE_URL'))
```

```python
supported_source_languages = ['C++', 'Python', 'Rust', 'Go', 'TypeScript', 'Node']
supported_target_languages = ['C++', 'Python', 'Rust', 'Go', 'TypeScript', 'Node']
supported_models = [
  "llama-3.3-70b-versatile",
  "llama-3.1-8b-instant",
  "openai/gpt-oss-20b",
  "gemma2-9b-it",
  "qwen-2.5-coder-32b"
]
```

```python
def convert(input_code, input_lang, output_lang, model):
  system = f'''You are a seasoned coder who has been tasked to convert the following code from {input_lang} to {output_lang}
  Make the code as efficient as possible, make no mistakes and output only and only the respective code, no other text or explanation.
  {input_lang} code-
  {input_code}
  '''
  user = f'''Conver the following code from {input_lang} to {output_lang}
  code-
  {input_code}'''

  message = [{
      "role": "system",
      "content": system
  }, {
      "role": "user",
      "content": user
  }]

  response = groq.chat.completions.create(model=model, messages= message)
  return response.choices[0].message.content
```

```python
with gr.Blocks() as ui:

  with gr.Row(equal_height=True):
    with gr.Column(scale = 6):
      input_lang = gr.Dropdown(supported_source_languages, value="C++", label="Input Language")
    with gr.Column(scale = 6):
      output_lang = gr.Dropdown(supported_target_languages, value="C++", label="Output Language")

  with gr.Row(equal_height=True):
    with gr.Column(scale = 6):
      input_code = gr.Code(label="Input Code")
    with gr.Column(scale = 6):
      output_code = gr.Code(label="Output Code")

  with gr.Row():
    with gr.Column():
      model = gr.Dropdown(label="Model", choices=supported_models)

  convert_btn = gr.Button("Convert")

  convert_btn.click(fn = convert,inputs=[input_code,input_lang, output_lang, model], outputs=[output_code])

ui.launch()
```
